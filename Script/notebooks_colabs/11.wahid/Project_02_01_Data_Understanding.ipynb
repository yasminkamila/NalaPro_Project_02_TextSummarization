```python
# Load Data
import pandas as pd

# Load ketiga split data
df_train = pd.read_json('/content/data_project2/train_text_clean.json')
df_dev   = pd.read_json('/content/data_project2/dev_text_clean.json')
df_test  = pd.read_json('/content/data_project2/test_text_clean.json')

print("Data berhasil dimuat!")
```

```python
for nama, df in [("TRAIN", df_train), ("DEV", df_dev), ("TEST", df_test)]:
    print(f"\n{'='*45}")
    print(f"  SPLIT: {nama}")
    print(f"{'='*45}")
    print(f"Jumlah baris & kolom : {df.shape}")
    print(f"\nNama kolom           : {list(df.columns)}")
    print(f"\nMissing values:")
    print(df.isnull().sum())

    # Cek duplikat hanya pada kolom 'id' karena kolom lain berisi nested list
    print(f"\nDuplikat (berdasarkan id) : {df['id'].duplicated().sum()} baris")

    print(f"\nContoh data (2 baris pertama):")
    display(df.head(2))
```
