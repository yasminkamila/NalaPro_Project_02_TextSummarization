```python
import pandas as pd
import re

# Load data
df_train = pd.read_json('/content/data_project2/train_text_clean.json')
df_dev = pd.read_json('/content/data_project2/dev_text_clean.json')
df_test = pd.read_json('/content/data_project2/test_text_clean.json')

print(f"Data siap dibersihkan! Total data latih: {len(df_train)}")
```

```python
def final_preprocessing(text):
    # 1. Pastikan string dan lowercase
    text = str(text).lower()

    # 2. Menghapus karakter non-alphanumeric (kecuali tanda baca dasar)
    text = re.sub(r'[^a-zA-Z0-9\s.,!?]', '', text)

    # 3. Menghapus spasi berlebih
    text = re.sub(r'\s+', ' ', text).strip()

    return text

# Terapkan ke semua split (Train, Dev, Test)
for df in [df_train, df_dev, df_test]:
    df['article_final'] = df['clean_article'].apply(final_preprocessing)
    df['summary_final'] = df['clean_summary'].apply(final_preprocessing)

print("Pembersihan akhir selesai!")
display(df_train[['article_final', 'summary_final']].head(2))
```

```python
# Buat folder output di Drive
output_path = '/content/drive/MyDrive/Colab Project/data_final'
if not os.path.exists(output_path):
    os.makedirs(output_path)

# Simpan ke JSON
df_train.to_json(f'{output_path}/train_final.json', orient='records', lines=False)
df_dev.to_json(f'{output_path}/dev_final.json', orient='records', lines=False)
df_test.to_json(f'{output_path}/test_final.json', orient='records', lines=False)

print(f"Berhasil! Data final sudah tersimpan rapi di: {output_path}")
```
